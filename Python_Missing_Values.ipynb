## Import pandas

```python
import pandas as pd
```

## Read and show data

```python
df = pd.read_csv('titanic.csv')
df.head()
```

```python
# show data information

df.info()
```

```python
# count mean value of Age

df['Age'].mean()
```

```python
# count median value of Age

df['Age'].median()
```

```python
# count mode value of Age

df['Age'].mode()
```

## Missing value

```python
# count the number of missing value in every column

df.isnull().sum()
```

```python
# count the number of missing value in persent

df.isnull().sum() / len(df) * 100.0
```

```python
# show all the rows with missing values in column Age

df[df['Age'].isnull() == True]
```

```python
# delete all the rows which contains missing values

delete_row = df.dropna(axis=0)
delete_row.info()
```

```python
# delete all the columns which contains missing values

delete_col = df.dropna(axis=1)
delete_col.info()
```

```python
# fill missing values in Age with value 20

df['Age'].fillna(20)
```

```python
# fill missing values in Age with mean value of Age
# also can use median(), min(), max(), etc

df['Age'].fillna(df['Age'].mean())
```

```python
# fill missing values with mode

df['Embarked'].fillna(df['Embarked'].mode()[0])
```

## Make a new dataframe that contains missing values

```python
df = pd.DataFrame([('Foreign Cinema', 'Resturant', pd.NA),
                  ('Liho liho', 'Restaurant', 224.0),
                  (pd.NA, 'bar', 80.5),
                  (pd.NA, 'bar', pd.NA),
                  (pd.NA, 'bar', 65.23),
                  ('Blue Barn', pd.NA, 361.98)],
          columns=('name', 'type', 'AvgBill')
                  )
df
```

```python
# fill missing values with backfill methods

df['name'].fillna(method='bfill')
```

```python
# fill missing values with frontfill methods

df['name'].fillna(method='ffill')
```
